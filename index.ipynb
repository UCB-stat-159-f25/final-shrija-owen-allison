## Project Overview

This project aims to utilize data science techniques to dissect these complex relationships, moving beyond assumptions to find statistically significant predictors of whether a student will graduate or drop out. Student retention and academic success are critical metrics for higher education institutions, influencing both university rankings and the economic prospects of students. By analyzing student data institutions can better understand the underlying causes of attrition. 

## Data Source

Our team retieved this dataset on early academic success from the UC Irvine Machine Learning Repository. The dataset, tited "Predict Students' Dropout and Academic Success", the dataset was constructed by consolidating several disjoint databases from a university. It contains features related to undergraduate students in fields such as agronomy, management, and nursing. The variables are divided into two categories: pre-enrollment features (academic path, demographics, socio-economic factors) and post-enrollment targets (academic performance in Semesters 1 and 2).
The dataset was already cleanly converted into a csv file, allowing us to simply store it in the folder 'data/'.

Link to Data Source: https://archive.ics.uci.edu/dataset/697 predict+students+dropout+and+academic+success

## Repository Structure

'data/:' contains the downloaded csv data file

'outputs/:' contains the generated plots and graphs

'tools/:' contains our tests

'main.ipynb:' Is our project notebook and provides an overview of the analysis and our results

'parent.ipynb:' Notebook containing data cleaning and preprocessing steps for the parent data

'student.ipynb:' Notebook containing data cleaning and preprocessing for the student data

'model_building.ipynb:' Notebook containing the logistic regression model training and evaluation 

'environment.yml:' Environment file with required packages for the project

'myst.yml:' Makefile to build JupyterBook for the repository and manage other tasks
