## Data Source

Our team retieved this dataset on early academic success from the UC Irvine Machine Learning Repository. The dataset, tited "Predict Students' Dropout and Academic Success", the dataset was constructed by consolidating several disjoint databases from a university. It contains features related to undergraduate students in fields such as agronomy, management, and nursing. The variables are divided into two categories: pre-enrollment features (academic path, demographics, socio-economic factors) and post-enrollment targets (academic performance in Semesters 1 and 2).
The dataset was already cleanly converted into a csv file, allowing us to simply store it in the folder 'data/'.
