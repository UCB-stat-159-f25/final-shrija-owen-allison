## Data Preprocessing

We first cleaned the data by only including columns that we would be analyzing, which included 'Admission grade', 'Target', and 'Tuition fees up to date', amognst others. We then removed the 794 students who were "Enrolled", as we had decided to focus on students who had already graduated or dropped out. After this we recategorized the 29 different parental "qualifications" into 12 new categories. After this, we decided to focus on 5 of the 12 categories, creating new data frames in the process.

## Exploratory Data Analysis

#### Parental Factors

Something that intrigued us when looking at this dataset were the two columns titled "Mother's qualificaton", and "Father's qualification", columns that told us both parent's education level. We at first hypothesized that the parent's education level would affect some student factors, however after doing some EDA and creating some figures, located in the parent_eda notebook, we saw that the parent's education level surprisingly didn't affect the student in terms of their graduation rate, tuition fees, debt, age at enrollment, or the qualification and admission grade. We only saw a small variation when examining scholarship rates, however the data wasn't strong enough to go ahead and conduct a logistic regression.

## Conclusion

In conclusion, we were left with three main findings: (1) parental factors were far less informative than we originally imagined in predicting student outcomes, (2) students age at enrollment, whether they're in debt, and whether they're up to date on tuition fees were the most significant student factors out of all we had access to, and finally (3) these aforementioned student factors are better at predicting if a student will go on to graduate rather than identifying students at risk of dropping out of school, and altogether are still insufficient by themselves if one is interested in creating a truly significant model for determing the chance a new student will go on to dropout or graduate.
