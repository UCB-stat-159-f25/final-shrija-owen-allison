## Conclusion

In conclusion, we were left with three main findings: (1) parental factors were far less informative than we originally imagined in predicting student outcomes, (2) students age at enrollment, whether they're in debt, and whether they're up to date on tuition fees were the most significant student factors out of all we had access to, and finally (3) these aforementioned student factors are better at predicting if a student will go on to graduate rather than identifying students at risk of dropping out of school, and altogether are still insufficient by themselves if one is interested in creating a truly significant model for determing the chance a new student will go on to dropout or graduate.
